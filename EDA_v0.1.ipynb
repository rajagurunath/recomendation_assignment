```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,15))
cnt=0
for i in range(0,2):
    for j in range(0,2):        
        ax[i][j].bar(df[cat_columns[cnt]].value_counts().index,df[cat_columns[cnt]].value_counts().values,label=cat_columns[cnt])
        ax[i][j].title.set_text(cat_columns[cnt])
        cnt+=1
```

```python
for i in cat_columns:
    df[i].value_counts().plot(kind='bar')
    plt.title(i)
    plt.show()
```
