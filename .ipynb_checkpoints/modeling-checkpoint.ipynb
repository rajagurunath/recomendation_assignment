```python
## Split Train-Test split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,2:],df['y_var'], test_size=0.33, random_state=5)
data_dict={'X_train':X_train,'X_test':X_test,'y_train':y_train,'y_test':y_test}
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
MODEL_DICT={
    'linear_reg':LinearRegression,
    'lasso':Lasso,
    'ridge':Ridge,
    'rf':RandomForestRegressor,
    'gbr':GradientBoostingRegressor,
    'cat_gbr':CatBoostRegressor,
    'light_gbr':LGBMRegressor,
}
```

```python
def train(model_dict,data_dict):
    trained_model_dict={}
    for name,model in model_dict.items():
        print(f"training ..{name}")
        trained_model_dict[name]=model().fit(data_dict['X_train'],data_dict['y_train'])
    return trained_model_dict


```

```python
trained_models=train(MODEL_DICT,data_dict)
```

```python
trained_models
```

```python
metrics_dict={
    "mae":mean_absolute_error,
    "mse":mean_squared_error,
    "rmse":lambda y_act,y_pred:mean_squared_error(y_act,y_pred)**0.5,
    "maae":median_absolute_error
}
```

```python
res_metrics=evaluate(trained_models,data_dict,metrics_dict=metrics_dict)
```

```python
def highlight_min(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.min()
    return ['background-color: green' if v else '' for v in is_max]
```

```python
res_metrics_df=pd.DataFrame(res_metrics)
ss=res_metrics_df.style.apply(highlight_min,axis=1)
```
