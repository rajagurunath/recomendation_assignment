```python
import shap
import catboost
import pandas as pd
import os
import numpy as np
np.set_printoptions(precision=4)
import catboost
from catboost import *
from catboost import datasets
from sklearn.model_selection import train_test_split
from pathlib import Path
```

```python
import joblib
```

```python
DIR=Path('../listing_price_suggest.csv')
DIR
df=pd.read_csv(DIR)
df.head()
```

```python
model=joblib.load('catboost_model.pkl')
```

```python
cat_columns=df.columns[df.columns.str.contains("cat")].tolist()
pool1 = Pool(data=df.iloc[:,2:], label=df['y_var'], cat_features=cat_columns)


shap_values = model.get_feature_importance(pool1,type='ShapValues')

expected_value = shap_values[0,-1]
shap_values = shap_values[:,:-1]

print(shap_values.shape)
```

```python
shap.initjs()
s=shap.force_plot(expected_value, shap_values[3,:], df.iloc[3,2:])
shap.save_html("single_exp.html",s)
s
```

```python
shap.initjs()
s=shap.force_plot(expected_value, shap_values[91,:], df.iloc[91,2:])
shap.save_html("single_exp1.html",s)
s
```

```python
shap.summary_plot(shap_values, df.iloc[:,2:])
```

```python
shap.summary_plot(shap_values, df.iloc[:,2:], plot_type="bar")
```

```python
X_small = df.iloc[0:2000,2:]
shap_small = shap_values[:2000]
s=shap.force_plot(expected_value, shap_small, X_small)
shap.save_html("explanation.html",s)
s
```
